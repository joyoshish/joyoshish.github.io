```python
from scipy.stats import binom
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
 
n, p = 10, 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf)
plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.show()
```
